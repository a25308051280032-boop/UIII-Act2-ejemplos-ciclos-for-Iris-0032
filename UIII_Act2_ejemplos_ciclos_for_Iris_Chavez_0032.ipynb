#### Ejemplos de ciclo for

```python
# Este programa muestra la tabla de multiplicar de un número usando un ciclo for

numero = int(input("Ingresa un número: ")) # Solicita un número y lo convierte a entero

for i in range(1, 11): # Recorre los números del 1 al 10
    resultado = numero * i # Multiplica el número ingresado por el valor de i
    print(numero, "x", i, "=", resultado) # Muestra el resultado de la multiplicación
```

```python
# Este programa cuenta cuántos números pares e impares ingresa el usuario

pares = 0 # Contador de pares
impares = 0 # Contador de impares

for i in range(5): # Repite el proceso 5 veces
    numero = int(input("Ingresa un número: ")) # Solicita un número

    if numero % 2 == 0: # Verifica si es par
        pares = pares + 1 # Incrementa contador de pares
    else: # Si no es par
        impares = impares + 1 # Incrementa contador de impares

print("Cantidad de números pares:", pares) # Muestra total de pares
print("Cantidad de números impares:", impares) # Muestra total de impares
```

```python
# Este programa verifica cuántas personas son mayores de edad

mayores = 0 # Contador de mayores de edad

for i in range(3): # Repite el proceso 3 veces
    edad = int(input("Ingresa una edad: ")) # Solicita la edad

    if edad >= 18: # Verifica si es mayor de edad
        print("Es mayor de edad") # Mensaje positivo
        mayores = mayores + 1 # Incrementa contador
    else:
        print("Es menor de edad") # Mensaje negativo

print("Total de mayores de edad:", mayores) # Resultado final
```

```python
# Este programa permite intentar ingresar una contraseña correcta

clave_correcta = "python123" # Contraseña guardada

for intento in range(3): # Permite 3 intentos

    clave = input("Ingresa la contraseña: ") # Solicita la contraseña

    if clave == clave_correcta: # Compara las contraseñas
        print("Acceso permitido") # Mensaje correcto
    else:
        print("Contraseña incorrecta") # Mensaje de error
```

```python
# Este programa solicita temperaturas y determina si son altas o bajas

altas = 0 # Contador de temperaturas altas
bajas = 0 # Contador de temperaturas bajas

for i in range(5): # Repite el proceso 5 veces

    temperatura = float(input("Ingresa una temperatura: ")) # Solicita la temperatura

    if temperatura >= 30: # Verifica si la temperatura es alta
        print("Temperatura alta") # Mensaje para temperaturas altas
        altas = altas + 1 # Incrementa el contador de altas

    else: # Si la condición no se cumple
        print("Temperatura baja") # Mensaje para temperaturas bajas
        bajas = bajas + 1 # Incrementa el contador de bajas

print("Cantidad de temperaturas altas:", altas) # Muestra total de altas
print("Cantidad de temperaturas bajas:", bajas) # Muestra total de bajas
```
